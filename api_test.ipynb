```python
import os
from dotenv import load_dotenv
import requests

#Load environment from root
load_dotenv('flask-arbitrage-app/.env')

API_KEY = os.getenv("ODDS_API_KEY")

# Base URL - replace with your actual API endpoint
BASE_URL = "https://api.the-odds-api.com/v3/odds/?apiKey=" + API_KEY

# Example: Test an API endpoint with apiKey in query params
def test_api_endpoint(endpoint: str, params: dict = None):
    """
    Test an API endpoint with apiKey authentication.
    
    Args:
        endpoint: The API endpoint path (e.g., "/odds", "/events")
        params: Optional query parameters
    
    Returns:
        Response JSON or error details
    """
    url = f"{BASE_URL}{endpoint}"
    
    # Add apiKey to query parameters
    query_params = {"apiKey": API_KEY}
    if params:
        query_params.update(params)
    
    try:
        response = requests.get(url, params=query_params, timeout=30)
        response.raise_for_status()
        return response.json()
    except requests.exceptions.RequestException as e:
        return {"error": str(e)}


# Example usage - uncomment and modify for your specific endpoint:
#result = test_api_endpoint("/odds", {"sport": "football"})
result = test_api_endpoint("&sport=baseball_mlb&region=us&mkt=totals")
print(result)
```
